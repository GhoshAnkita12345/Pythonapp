# **DATASET CONNECTIVITY**

```python
import mysql.connector
import pandas as pd

connection = mysql.connector.connect(
user='root',
password='root',
host='localhost',
database='ecommerce1'
)
# Create a cursor object to execute SQL queries
cursor = connection.cursor()
# Query data from the 'customer' table
cursor.execute('SELECT * FROM customer')
#After fetching data from the database we are storing it into Pandas DataFrame
customer_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in
cursor.description])
# Query data from the 'product' table
cursor.execute('SELECT * FROM product')
product_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in
cursor.description])
# Query data from the 'order_details' table
cursor.execute('SELECT * FROM order_details')
order_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in
cursor.description])
#printing first 5 records from each table
print("Customer Data:")
print(customer_data.head())
print("\nProduct Data:")
print(product_data.head())
print("\nOrder Data:")
print(order_data.head())
```

# **DATA PROCESSING AND DISPLAY**

## FOR CUSTOMERS DATA:

```python
# Check for Missing Values
print("Missing Values in Customer Database:")
print(customer_df.isnull().sum())

# Check for Duplicates
print("\nDuplicate Rows in Customer Database:")
print(customer_df.duplicated().sum())
customer_df.drop_duplicates(inplace=True)

# Display Cleaned Data
print("\nCleaned Customer Database:")
print(customer_df)
```

## FOR PRODUCTS DATA:

```python
# Check for Missing Values
print("\nMissing Values in Product Database:")
print(product_df.isnull().sum())

# Check for Duplicates
print("\nDuplicate Rows in Product Database:")
print(product_df.duplicated().sum())
product_df.drop_duplicates(inplace=True)

# Display Cleaned Data
print("\nCleaned Product Database:")
print(product_df)
```

## FOR ORDERS DATA:

```python
# Check for Missing Values
print("\nMissing Values in Order Database:")
print(order_df.isnull().sum())

# Check for Duplicates
print("\nDuplicate Rows in Order Database:")
print(order_df.duplicated().sum())
order_df.drop_duplicates(inplace=True)

# Display Cleaned Data
print("\nCleaned Order Database:")
print(order_df)
```

# **EXPLORING ECOMMERCE THROUGH VISUALIZATION**

## Visualization 1: Bar chart of City-wise Customer Count

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

city_counts = customer_df['city'].value_counts()
city_counts.plot(kind='bar', figsize=(10, 6))
plt.title('City-wise Customer Count')
plt.xlabel('City')
plt.ylabel('Customer Count')
plt.xticks(rotation=45)
plt.show()
```

## Visualization 2: Pie chart of Payment Modes Distribution

```python

payment_modes_counts = order_df['payment_mode'].value_counts()
payment_modes_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, figsize=(8, 8))
plt.title('Payment Modes Distribution')
plt.ylabel('')
plt.show()
```

## Visualization 3: Box plot of Selling Prices of Products

```python
product_df['selling_price'].plot(kind='box', figsize=(8, 6))
plt.title('Distribution of Selling Prices')
plt.ylabel('Selling Price')
plt.show()
```

## Visualization 4: Histogram of Order Quantities

```python
order_df['quantity'].plot(kind='hist', color='skyblue', bins=20, figsize=(8, 6))
plt.title('Distribution of Order Quantities')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

## Visualization 5: Bar chart of Category-wise Product Count

```python
category_counts = product_df['category'].value_counts()
category_counts.plot(kind='bar', color='lightcoral', figsize=(10, 6))
plt.title('Category-wise Product Count')
plt.xlabel('Category')
plt.ylabel('Product Count')
plt.xticks(rotation=45)
plt.show()
```

## Visualization 6: Line plot of Order Date vs. Total Price

```python
order_df['order_date'] = pd.to_datetime(order_df['order_date'])
order_df.plot(x='order_date', y='total_price', figsize=(12, 6))
plt.title('Order Date vs. Total Price')
plt.xlabel('Order Date')
plt.ylabel('Total Price')
plt.xticks(rotation=45)
plt.show()
```

## Visualization 7: Scatter plot of Original Price vs. Selling Price

```python
plt.scatter(product_df['original_price'], product_df['selling_price'])
plt.title('Original Price vs. Selling Price')
plt.xlabel('Original Price')
plt.ylabel('Selling Price')
plt.show()
```

## Visualization 8: Bar chart of Sub-category-wise Product Count

```python
sub_category_counts = product_df['sub_category'].value_counts()
sub_category_counts.plot(kind='bar', color= 'coral', figsize=(10, 6))
plt.title('Sub-category-wise Product Count')
plt.xlabel('Sub-category')
plt.ylabel('Product Count')
plt.xticks(rotation=45)
plt.show()
```

## Visualization 9: Pie chart of Order Status Distribution

```python
order_status_counts = order_df['order_status'].value_counts()
order_status_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, figsize=(8, 8))
plt.title('Order Status Distribution')
plt.ylabel('')
plt.show()
```

## Visualization 10: Bar chart of top 10 selling products based on the quantity sold

```python
top_product = order_df['product_id'].value_counts().head(10)
top_product_df = order_df[order_df['product_id'].isin(top_product.index)]

# Group by product_id and sum the quantities sold
top_products_quantity = top_product_df.groupby('product_id')['quantity'].sum().sort_values(ascending=False)

# Get product names for the top selling products
top_product_names = product_df.set_index('product_id').loc[top_products_quantity.index]['product_name']

# Plot the bar chart
top_products_quantity.plot(kind='bar', figsize=(12, 6))
plt.title('Top 10 Selling Products')
plt.xlabel('Product Name')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45, ha='right')
plt.show()
```
