# DATASET INFORMATION

Healthcare-data-stroke-data.csv: The csv contains data related to patients who may have heart disease and various attributes which determine that :

* **id:** unique identifier

- **gender:** "Male", "Female" or "Other"

- **age:** age of the patient

- **hypertension:** 0 if the patient doesn't have hypertension, 1 if the patient has hypertension

- **heart_disease:** 0 if the patient doesn't have any heart diseases, 1 if the patient has a heart disease

- **ever_married:** "No" or "Yes"

- **work_type:** "children", "Govt_jov", "Never_worked", "Private" or "Self-employed"

- **Residence_type:** "Rural" or "Urban"

- **avg_glucose_level:** average glucose level in blood

- **bmi:** body mass index

- **smoking_status:** "formerly smoked", "never smoked", "smokes" or "Unknown"*

- **stroke:** 1 if the patient had a stroke or 0 if not *Note: "Unknown" in smoking_status means that the information is unavailable for this patient

# DATA CLEANING PROCESS:

## 1.HANDLING MISSING VALUES

```python
import pandas as pd

# Load the dataset
file_path = "healthcare-dataset-stroke-data.csv"
stroke_data = pd.read_csv(file_path)

# Check for missing values
missing_values = stroke_data.isnull().sum()
print("Missing values before handling:")
print(missing_values)

# Handle missing values by filling with the mean
stroke_data['bmi'] = stroke_data['bmi'].fillna(stroke_data['bmi'].mean())

# Confirm missing values have been handled
missing_values_after = stroke_data.isnull().sum()
print("\nMissing values after handling:")
print(missing_values_after)

# Store the cleaned dataset into a new file
cleaned_file_path = "cleaned_healthcare_dataset.csv"
stroke_data.to_csv(cleaned_file_path, index=False)
print(f"\nCleaned dataset saved to file: {cleaned_file_path}")
```

## 2.REMOVING DUPLICATE VALUES:

```python
import pandas as pd

# Load the dataset
file_path = "healthcare-dataset-stroke-data.csv"
stroke_data = pd.read_csv(file_path)

# Remove duplicate rows and columns
stroke_data_cleaned = stroke_data.T.drop_duplicates().T

# Display the cleaned dataset
print("\nDataset after removing duplicate values:")
print(stroke_data_cleaned)

# Save the cleaned dataset to a new CSV file 
cleaned_file_path = "cleaned_healthcare_dataset.csv"
stroke_data_cleaned.to_csv(cleaned_file_path, index=False)
print(f"\nCleaned dataset saved to file: {cleaned_file_path}")
```

## 1) Does age has impact on strokes? and How is this parameter distributed?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "new_cleaned_healthcare_dataset.csv"
stroke_data = pd.read_csv(file_path) 

# Filter data for patients who have had a stroke
stroke_patients = stroke_data[stroke_data['stroke'] == 1]

# Create a histogram to visualize the distribution of ages among stroke patients
plt.figure(figsize=(10, 6))
plt.hist(stroke_patients['age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages Among Stroke Patients')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### INFERENCE: 

The histogram displays the distribution of ages among patients who have had a stroke.
Observing the distribution of ages can provide insights into whether age impacts the likelihood of experiencing a stroke.
Analyzing the distribution may reveal any trends or patterns in stroke incidence across different age groups.
Further statistical analysis may be required to determine the precise impact of age on strokes and to identify any potential age-related risk factors for stroke.





## 2) Does body mass index and glucose levels in a person, propel a heart stroke?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "new_cleaned_healthcare_dataset.csv"
stroke_data = pd.read_csv(file_path)

# Data Visualization
plt.figure(figsize=(10, 6))
plt.scatter(stroke_data['bmi'], stroke_data['avg_glucose_level'], c=stroke_data['stroke'], cmap='coolwarm', alpha=0.6)
plt.colorbar(label='Stroke Status')
plt.xlabel('BMI')
plt.ylabel('Average Glucose Level')
plt.title('Relationship between BMI, Glucose Levels, and Stroke')
plt.grid(True)
plt.show()

# Statistical Analysis
stroke_group = stroke_data.groupby('stroke')
stroke_yes = stroke_group.get_group(1)['bmi']
stroke_no = stroke_group.get_group(0)['bmi']

# Compute mean BMI for individuals with and without stroke
mean_bmi_yes = stroke_yes.mean()
mean_bmi_no = stroke_no.mean()

# Print the mean BMI for individuals with and without stroke
print("Mean BMI for individuals with stroke:", mean_bmi_yes)
print("Mean BMI for individuals without stroke:", mean_bmi_no)
```

### INFERENCE:

From the scatter plot, we observe that there doesn't seem to be a clear linear relationship between BMI, glucose levels, and stroke occurrence.
The mean BMI for individuals with a stroke and those without a stroke can be compared to see if there's a significant difference.
However, this preliminary analysis does not provide conclusive evidence on whether BMI and glucose levels directly cause strokes. Further analysis using more advanced statistical techniques may be necessary to draw robust conclusions.


## 3) Is there a difference in the rate of heart stroke for smokers and non smokers?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "new_cleaned_healthcare_dataset.csv"
stroke_data = pd.read_csv(file_path)

# Calculate the proportion of stroke cases among smokers and non-smokers
smokers_stroke_rate = stroke_data[stroke_data['smoking_status'] == 'smokes']['stroke'].mean()
non_smokers_stroke_rate = stroke_data[stroke_data['smoking_status'] == 'never smoked']['stroke'].mean()

# Create a bar chart to visualize stroke rates among smokers and non-smokers
plt.figure(figsize=(8, 6))
plt.bar(['Smokers', 'Non-Smokers'], [smokers_stroke_rate, non_smokers_stroke_rate], color=['skyblue', 'salmon'])
plt.xlabel('Smoking Status')
plt.ylabel('Stroke Rate')
plt.title('Stroke Rate Among Smokers and Non-Smokers')
plt.ylim(0, 0.2)  # Set y-axis limits for better visualization
plt.show()
```

### INFERENCE:

The bar chart clearly illustrates the difference in stroke rates between smokers and non-smokers.
If the stroke rate among smokers is significantly higher than non-smokers, it suggests that smoking may be a risk factor for stroke.
Conversely, if the stroke rate among non-smokers is higher, it indicates that other factors might have a more significant impact on stroke risk for non-smokers.
This visualization provides valuable insights into the relationship between smoking status and stroke risk, which can inform public health interventions and smoking cessation programs.

## 4) Is there a relationship between the type of work (private sector, self-employed, etc.) and the risk of experiencing a stroke?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "new_cleaned_healthcare_dataset.csv"
stroke_data = pd.read_csv(file_path)

# Filter out missing values in the 'work_type' column
stroke_data.dropna(subset=['work_type'], inplace=True)

# Create a box plot to visualize the distribution of stroke cases by work type
plt.figure(figsize=(10, 6))
stroke_data.boxplot(column='age', by='work_type', grid=False, vert=False)
plt.xlabel('Age')
plt.ylabel('Work Type')
plt.title('Distribution of Age by Work Type')
plt.suptitle('')
plt.show()
```

### INFERENCE:

The box plot showcases the distribution of ages across various types of work, such as private sector, self-employed, etc.
By comparing the distributions of ages within each work type, we can infer potential relationships between the type of work and the risk of experiencing a stroke.
Observing any differences in the median age or spread of ages across different work types may provide insights into the association between occupation and stroke risk.
Further analysis and consideration of other factors may be necessary to ascertain the nature and significance of the relationship between work type and stroke risk.

## 5) How does hypertension status correlate with the occurrence of strokes?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "new_cleaned_healthcare_dataset.csv"
stroke_data = pd.read_csv(file_path)

# Calculate the proportion of stroke cases among individuals with and without hypertension
hypertension_stroke_count = stroke_data[stroke_data['hypertension'] == 1]['stroke'].value_counts()
no_hypertension_stroke_count = stroke_data[stroke_data['hypertension'] == 0]['stroke'].value_counts()

# Create pie charts to visualize stroke rates among individuals with and without hypertension
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart for individuals with hypertension
axs[0].pie(hypertension_stroke_count, labels=hypertension_stroke_count.index, autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
axs[0].set_title('Hypertension Status: With Hypertension')

# Pie chart for individuals without hypertension
axs[1].pie(no_hypertension_stroke_count, labels=no_hypertension_stroke_count.index, autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
axs[1].set_title('Hypertension Status: Without Hypertension')

plt.show()
```

### INFERENCE:

The pie charts provide a visual representation of the proportion of stroke cases among individuals with and without hypertension.
By comparing the sizes of the slices in each pie chart, we can observe the relative contribution of strokes in each group.
If a larger proportion of strokes occurs in individuals with hypertension compared to those without hypertension, it suggests a correlation between hypertension status and stroke occurrence.
Conversely, if the proportion of strokes is similar between the two groups, other factors may play a more significant role in stroke risk.





## 6) Is there a relationship between heart disease and the likelihood of experiencing a stroke?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "new_cleaned_healthcare_dataset.csv"
stroke_data = pd.read_csv(file_path)

# Filter data for individuals with and without heart disease
heart_disease_data = stroke_data[stroke_data['heart_disease'] == 1]
no_heart_disease_data = stroke_data[stroke_data['heart_disease'] == 0]

# Create scatter plots for age vs. stroke occurrence for individuals with and without heart disease
plt.figure(figsize=(10, 6))

plt.scatter(heart_disease_data['age'], heart_disease_data['stroke'], color='red', label='With Heart Disease', alpha=0.5)
plt.scatter(no_heart_disease_data['age'], no_heart_disease_data['stroke'], color='blue', label='Without Heart Disease', alpha=0.5)

plt.xlabel('Age')
plt.ylabel('Stroke Occurrence')
plt.title('Age vs. Stroke Occurrence by Heart Disease Status')
plt.legend()
plt.grid(True)

plt.show()
```

### INFERENCE:

In the scatter plot, each point represents an individual's age and stroke occurrence.
The red points represent individuals with heart disease, while the blue points represent those without heart disease.
By observing the distribution and clustering of points, we can infer whether individuals with heart disease are more likely to experience strokes compared to those without heart disease.
If there is a clear clustering of strokes among individuals with heart disease at certain ages, it suggests a stronger relationship between heart disease and stroke occurrence.

## 7) Does the residential type (urban or rural) influence the incidence of strokes?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "new_cleaned_healthcare_dataset.csv"
stroke_data = pd.read_csv(file_path)

# Filter data for urban and rural areas
urban_stroke_data = stroke_data[stroke_data['Residence_type'] == 'Urban']
rural_stroke_data = stroke_data[stroke_data['Residence_type'] == 'Rural']

# Create histograms for stroke incidence in urban and rural areas
plt.figure(figsize=(10, 6))
plt.hist(urban_stroke_data['stroke'], bins=2, color='skyblue', alpha=0.7, label='Urban')
plt.hist(rural_stroke_data['stroke'], bins=2, color='orange', alpha=0.7, label='Rural')
plt.xlabel('Stroke Incidence')
plt.ylabel('Frequency')
plt.title('Distribution of Stroke Incidence by Residential Type')
plt.xticks([0.25, 0.75], ['No Stroke', 'Stroke'])
plt.legend()
plt.show()
```

### INFERENCE:

The violin plots provide insight into the distribution of stroke cases between urban and rural residential types.
Comparing the shapes and spreads of the violin plots may reveal potential differences in stroke incidence between urban and rural areas.
However, further statistical analysis is necessary to determine the significance of any observed differences and to account for potential confounding variables.

## 8) What is the distribution of BMI among patients who have had a stroke compared to those who haven't?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "new_cleaned_healthcare_dataset.csv"
stroke_data = pd.read_csv(file_path)

# Filter data for patients who have had a stroke and those who haven't
stroke_patients = stroke_data[stroke_data['stroke'] == 1]
non_stroke_patients = stroke_data[stroke_data['stroke'] == 0]

# Create KDE plots to visualize the distribution of BMI among stroke and non-stroke patients
plt.figure(figsize=(10, 6))

# KDE plot for stroke patients
plt.subplot(1, 2, 1)
plt.hist(stroke_patients['bmi'], bins=20, density=True, alpha=0.6, color='red')
plt.title('BMI Distribution for Stroke Patients')
plt.xlabel('BMI')
plt.ylabel('Density')

# KDE plot for non-stroke patients
plt.subplot(1, 2, 2)
plt.hist(non_stroke_patients['bmi'], bins=20, density=True, alpha=0.6, color='blue')
plt.title('BMI Distribution for Non-Stroke Patients')
plt.xlabel('BMI')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

### INFERENCE:

There might be a correlation between BMI and the occurrence of strokes, as observed by potential differences in the BMI distributions between stroke and non-stroke patients.
Higher BMI values among stroke patients could suggest a potential risk factor for stroke occurrence.
Further statistical analysis, such as hypothesis testing or regression modeling, can provide deeper insights into the relationship between BMI and strokes and help identify potential preventive measures or interventions for stroke risk management.





## 9) Are there any notable trends in glucose levels among patients who have had a stroke?

## 10) How does gender impact the prevalence of strokes?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "new_cleaned_healthcare_dataset.csv"
stroke_data = pd.read_csv(file_path)

# Count the number of stroke cases by gender
stroke_by_gender = stroke_data.groupby('gender')['stroke'].sum()

# Create a pie chart to visualize the distribution of stroke cases by gender
plt.figure(figsize=(8, 6))
plt.pie(stroke_by_gender, labels=stroke_by_gender.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Distribution of Stroke Cases by Gender')
plt.show()
```

### INFERENCE:

The pie chart illustrates the proportion of stroke cases among males and females.
We can observe the relative prevalence of strokes between genders and assess if there's a notable difference.
Further statistical analysis could explore whether gender significantly impacts the likelihood of experiencing a stroke.
Understanding the gender-specific patterns of stroke prevalence can help tailor preventive strategies and healthcare interventions to address potential disparities.
